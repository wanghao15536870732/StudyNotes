## %timeit

```python
%timeit L = [i**2 for i in range(1000)]
```

```python
%timeit L = [i**2 for i in range(100000)]
```

```python
%timeit L = [i**2 for i in range(10)]
```

```python
%%timeit
L = []
for n in range(1000):
    L.append(n ** 2)
```
