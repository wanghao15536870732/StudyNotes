### scili-learn中的accuracy_score

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split # 导入我们的方法
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2) # 获取我们的训练跟测试数据集
```

```python
from sklearn.neighbors import KNeighborsClassifier # 封装的kNN算法

knn_clf = KNeighborsClassifier(n_neighbors=3) # 保存
X_train.shape
```

```python
y_train.shape
```

```python
knn_clf.fit(X_train,y_train) # fit数据
```

```python
y_predict = knn_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score  # 导入计算准确率的函数
accuracy_score(y_test,y_predict)
```

```python
knn_clf.score(X_test,y_test) # 直接查看准确率
```
