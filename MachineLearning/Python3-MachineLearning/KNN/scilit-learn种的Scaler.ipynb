### scilit-learnScaler

```python
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris() # 
```

```python
X = iris.data # 
y = iris.target # 
```

```python
X[:10,:] # 10
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_Train,X_test,y_train,y_test = train_test_split(iris.data,iris.target,random_state=666) # 
```

### scilit-learnStandardScaler  

```python
from sklearn.preprocessing import StandardScaler # import
```

```python
standardScaler = StandardScaler()
standardScaler.fit(X_Train) # 
```

```python
standardScaler.mean_ # 
```

```python
standardScaler.scale_ # ,
```

```python
standardScaler.transform(X_Train) # 
```

```python
X_Train = standardScaler.transform(X_Train) # x_train
```

```python
X_Train
```

```python
standardScaler.transform(X_test)
```

```python
X_test_standerd = standardScaler.transform(X_test) # 
```

```python
X_test_standerd
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_clf = KNeighborsClassifier(n_neighbors=3) # 
```

```python
knn_clf.fit(X_Train,y_train)
```

```python
knn_clf.score(X_test_standerd,y_test) # 
```

```python
knn_clf.score(X_test,y_test) # 
```
