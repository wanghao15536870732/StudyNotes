## 其他创建 numpy.array的方法

```python
np.zeros(10)  # 全为0的大小10的数组
```

```python
np.zeros(10).dtype  # 默认浮点型
```

```python
np.zeros(10,dtype=int)
```

```python
np.zeros((3,5)) # 传入二维的元组，生成矩阵
```

```python
np.zeros(shape=(3,5),dtype=int) # 参数
```

```python
np.ones(10)  # 生成全1的向量
```

```python
np.ones((3,5))  # 全1矩阵
```

```python
np.full(shape=(3,5),fill_value=666.0)  # full()方法生成指定数据矩阵，.0默认浮点型
```

### arange方法

```python
[i for i in range(0,20,2)] 
```

```python
np.arange(0,20,2)  # 与range()参数意义相同
```

```python
np.arange(0,1,0.2)  # arange()可以接受步长为0.2的数组
```

```python
np.arange(0,10)  # 默认步长为1
```

```python
np.arange(10)   # 默认从0开始
```

### linspace 方法

```python
np.linspace(0,20,10) # 参数与arange()相同，10表示等长的截出10个元素，包括0跟20
```

```python
np.linspace(0,20,11) # 一共需要截出11个数
```

### random 方法  生成随机数

```python
np.random.randint(0,10)  # 0-10之间的随机数
```

```python
np.random.randint(0,10,10) #生0-10的大小为10的随机数组，不包含10本身，无法取到10
```

```python
np.random.randint(0,1,10) # 不包含1本身
```

```python
np.random.randint(4,8,size=(3,5)) # 生成二维随机矩阵
```

```python
np.random.randint(4,8,size=(3,5))  #与上述不同
```

```python
np.random.seed(666)  # 设定种子  固定随机数
```

```python
np.random.randint(4,8,size=(3,5)) 
```

```python
np.random.seed(666)
np.random.randint(4,8,size=(3,5))   #随机种子指定随机数
```

```python
np.random.random()  # 默认生成01之间的浮点数
```

```python
np.random.random(10) # 指定10个数据，随机向量  默认均匀分布
```

```python
np.random.random((3,5)) # 生成0-1之间的[3,5]随机浮点矩阵   默认均匀分布
```

### normal  方法  正态分布随机数

```python
np.random.normal()  #均值为0，方差为1 的随机浮点数
```

```python
np.random.normal(10,100) # 均值为10，方差为100 的随机浮点数
```

```python
np.random.normal(0,1,(3,5)) # 均值为0，方差为1 的随机浮点[3,5]矩阵数据
```
