### 数据加载 & 数据探索

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris.keys() # 查看特征组合
```

```python
print(iris.DESCR) # 查看具体内容
```

```python
iris.data # 数据
```

```python
iris.data.shape # 查看数据的形状
```

```python
iris.feature_names # 四列数据代表的意思
```

```python
iris.target # 鸢尾花的类型
```

```python
iris.target.shape
```

```python
iris.target_names # 鸢尾花的类型
```

```python
X = iris.data[:,:2] # 取得前两列的数据，即鸢尾花的sepal length 和 sepal width
X.shape
```

```python
plt.scatter(X[:,0],X[:,1]) # 取出第0列为横坐标 取出第1列为纵坐标
```

```python
y = iris.target # 取出每个数据的鸢尾花的类型向量
```

```python
plt.scatter(X[y==0,0],X[y==0,1],color="red") # 第一种鸢尾花
plt.scatter(X[y==1,0],X[y==1,1],color="blue") # 第二种鸢尾花
plt.scatter(X[y==2,0],X[y==2,1],color="green") # 第三种鸢尾花
plt.show()

# 第一种跟其他鸢尾花的sepal length 和 sepal width 更能区分出他们
```

```python
plt.scatter(X[y==0,0],X[y==0,1],color="red",marker="o") # 第一种鸢尾花 ,marker指定样式
plt.scatter(X[y==1,0],X[y==1,1],color="blue",marker="+") # 第二种鸢尾花
plt.scatter(X[y==2,0],X[y==2,1],color="green",marker="x") # 第三种鸢尾花
plt.show()
```

```python
X = iris.data[:,2:] # 取得后两列，鸢尾花的petal length 和 petal width
```

```python
plt.scatter(X[y==0,0],X[y==0,1],color="red",marker="o") # 第一种鸢尾花 ,marker指定样式
plt.scatter(X[y==1,0],X[y==1,1],color="blue",marker="+") # 第二种鸢尾花
plt.scatter(X[y==2,0],X[y==2,1],color="green",marker="x") # 第三种鸢尾花
plt.show()

# 最终表明 第二种跟第三种鸢尾花的petal length 和 petal width 更能区分出他们
```
