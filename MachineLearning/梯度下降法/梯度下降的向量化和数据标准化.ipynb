# 梯度下降法的向量化

### $ \frac2m\cdot \left\lgroup\begin{matrix} \sum_{i=1}^{m}{({X_b^{(i)\theta - y^{(i)}}) \cdot{X_0^{(i)}} }}  \cr \sum_{i=1}^{m}{({X_b^{(i)\theta - y^{(i)}}) \cdot{X_1^{(i)}} }} \cr \sum_{i=1}^{m}{({X_b^{(i)\theta - y^{(i)}}) \cdot{X_2^{(i)}} }} \cr ... \cr \sum_{i=1}^{m}{({X_b^{(i)\theta - y^{(i)}}) \cdot{X_n^{(i)}} }} \end{matrix}\right\rgroup 可化简转换为矩阵的乘法$
### $ = \frac2m\cdot \left[\begin{matrix} X_b^{(1)}\theta - y^{(1)}, & X_b^{(2)}\theta - y^{(2)}, & X_b^{(3)}\theta - y^{(3)}, & ... & X_b^{(m)}\theta - y^{(m)} \end{matrix}\right] \cdot \left[\begin{matrix} X_0^{(1)} & X_1^{(1)} & X_2^{(1)} & ... & X_n^{(1)} \cr X_0^{(2)} & X_1^{(2)} & X_2^{(2)} & ... & X_n^{(2)} \cr X_0^{(3)} & X_1^{(3)} & X_2^{(3)} & ... & X_n^{(3)}\cr ... & ... & ... & ... & ... \cr X_0^{(m)} & X_1^{(m)} & X_2^{(m)} & ... & X_n{(m)} \end{matrix}\right]$
## $ = \frac2m \cdot (X_b\theta - y)^{T} \cdot X_b  = \frac2m \cdot X_b^{T} \cdot (X_b\theta - y)$ 

```python
# 化简改进先前的梯度求法
def dJ(theta, X_b, y):
    # res = np.empty(len(theta))
    # res[0] = np.sum(X_b.dot(theta) - y)
    # for i in range(1, len(theta)):
    #     res[i] = (X_b.dot(theta) - y).dot(X_b[:, i])
    # return res * 2 / len(X_b)
    return X_b.T.dot(X_b.dot(theta) - y) * 2 / len(X_b)
```

```python
import numpy as np
from sklearn import datasets
```

```python
boston = datasets.load_boston()
X = boston.data
y = boston.target

X = X[y < 50.0]
y = y[y < 50.0]
```

```python
from model_selection import train_test_split

X_train,y_train,X_test,y_test = train_test_split(X,y,seed=666)
```

```python
from LinearRegression import LinearRegression
lin_reg1 = LinearRegression()
%time lin_reg1.fit_normal(X_train,y_train)
lin_reg1.score(X_test,y_test)
```

### 使用梯度下降法

```python
lin_reg2 = LinearRegression()
lin_reg2.fit_gd(X_train,y_train)
```

```python
lin_reg2.coef_ #没有乘2/m导致 result too large
```

```python
X_train[:10,:]
```

```python
lin_reg2.fit_gd(X_train,y_train,eta=0.000001)
```

```python
lin_reg2.score(X_test,y_test)
```

```python
%time lin_reg2.fit_gd(X_train,y_train,eta=0.000001,n_iters=1e6)
```

```python
lin_reg2.score(X_test,y_test) # 不同维度导训练精度下降
```

## 梯度下降法的优势

```python
m = 1000
n = 5000

big_x = np.random.normal(size=(m,n))
true_theta = np.random.uniform(0.0,100.0,size=n+1)
big_y = big_x.dot(true_theta[1:]) + true_theta[0] + np.random.normal(0.,10.,size=m)
```

```python
big_reg1 = LinearRegression() # 实例化线性回归模型
%time big_reg1.fit_normal(big_x,big_y)# 使用正规方程，耗时比较明显
```

```python
big_reg2 = LinearRegression()
%time big_reg2.fit_gd(big_x,big_y) # 使用梯度下降法方程，运算速度比较快
```
